```python
def area_triangolo(a,b):
    area = (a*b)/2
    print("l'area del triangolo è " + str(area))
```

```python
area_triangolo(5,2)
```

```python
def area_rettangolo(a,b):
    area = a*b
    print("l'area del rettangolo è " +str(area))
```

```python
area_rettangolo(10,3)
```

```python
def area_quadrato(a):
    area = a**2
    print("l'area del quadrato è " +str(area))
```

```python
class Persona:
    def __init__(self,a,b):
        self.nome = a
        self.eta = b
    def double(self,x):
        return x*2
    def triple(self,x):
        return x*3
    def quadratoanni(self):
        return self.eta**2
pietro = Persona('pietro',28)
```

```python
pietro.quadratoanni()
```
